```python
import arcpy
import pandas as pd
import os

def count_pixels_for_all_values(raster_path, class_name, output_csv):
    arcpy.env.addOutputsToMap = False

    arcpy.management.BuildRasterAttributeTable(
    in_raster=raster_path,
    overwrite="Overwrite",
    convert_colormap="NONE"
    )    
    try:
        # Check if the raster has the required fields
        fields = [f.name for f in arcpy.ListFields(raster_path)]
        
        if class_name not in fields:
            arcpy.AddWarning(f"Field '{class_name}' not found in {os.path.basename(raster_path)}. Skipping.")
            return
            
        if "Count" not in fields:
            arcpy.AddWarning(f"Field 'Count' not found in {os.path.basename(raster_path)}. (Floating-point rasters do not have attribute tables). Skipping.")
            return
        
        # Query the table for all classes
        pixel_count = []
        with arcpy.da.SearchCursor(raster_path, [class_name, "Count"]) as cursor:
            for row in cursor:
                pixel_count.append({
                    "Raster": os.path.basename(raster_path), 
                    "Class": row[0], 
                    "Pixel Count": row[1]
                })
        
        # Convert results to a DataFrame
        df = pd.DataFrame(pixel_count)
        
        # Append results to CSV file
        if not df.empty:
            df.to_csv(output_csv, mode='a', header=not os.path.exists(output_csv), index=False)
            arcpy.AddMessage(f"Success: Exported pixel counts for {os.path.basename(raster_path)}")
        else:
            arcpy.AddWarning(f"No pixel data found in {os.path.basename(raster_path)}")
            
    except Exception as e:
        # This will now correctly push errors to the ArcGIS Pro message window
        arcpy.AddError(f"Error processing raster {os.path.basename(raster_path)}: {str(e)}")
```
